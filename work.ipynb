# Step 2: Data Preparation
- Droping irrelevant columns and rows
- Identifying duplicated columns
- Renaming Columns
- Feature Creation

```python
##SUBSET THE DATASET TO RETAIN SELECTED VARIABLES AND DROP UNNECESSARY COLUMNS##
df = df[['coaster_name', 
    # 'Length', 'Speed',
     'Location', 'Status', 
    #  'Opening date','Type', 
    'Manufacturer', 
    # 'Height restriction', 'Model', 'Height',
    #    'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #    'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #    'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #    'Track layout', 'Fastrack available', 'Soft opening date.1',
    #    'Closing date', 'Opened', 'Replaced by', 'Website',
    #    'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #    'Single rider line available', 'Restraint Style',
    #    'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
        #  'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
        #  'height_value', 'height_unit',
           'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
##CHANGE DATATYPE TO DATE##
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
##RENAME COLUMN
df = df.rename(columns= {'coaster_name' : 'Coaster_Name',
                    'year_introduced' : 'Year_Introduced',
                    'opening_date_clean' : 'Opening_Date',
                    'speed_mph' : 'Speed_mph',
                    'height_ft' : 'Height_ft',
                    'Inversions_clean' : 'Inversions',
                    'Gforce_clean' : 'Gforce'})
```

```python
##CHECK HOW MANY NAN VALUE IN EACH COLUMN##
df.isna().sum()
```

```python
##EXTRACT DUPLICATE ROWS FROM THE DATASET##
df.loc[df.duplicated()]
```

```python
##REMOVE DUPLICATE COASTER RECORD BASED ON NAME, LOCATION AND OPENING DATE##
df = df.loc[~df.duplicated(subset= ['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True).copy()
```

# Ask a Question about the data

- Try answer a question you have about the data using a plot or statistic:
    - What are the locations with the fastest roller coaster (minimum of 10)?

```python
df.query('Location != "Other"')\
    .groupby('Location')['Speed_mph']\
    .agg(['mean', 'count'])\
    .query ('count >= 10')\
    .sort_values('mean')
```

```python
ax = df.query('Location != "Other"')\
    .groupby('Location')['Speed_mph']\
    .agg(['mean', 'count'])\
    .query ('count >= 10')\
    .sort_values('mean')['mean']\
    .plot(kind = 'barh', figsize = (12, 5), title = 'Avearge Coaster Speed by Location')

ax.set_xlabel('Average Coaster Speed')
plt.show()
```
